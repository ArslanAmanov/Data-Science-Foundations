# Change Payer Name as PLT  

```python
def payer(txt):
    try:
        if txt == 'บริษัท พรูเด็นเชียล ประกันชีวิต (ประเทศไทย) จำกัด (มหาชน)' or 'บริษัท พรูเด็นเชียล ประกันชีวิต (ประเทศไทย) จำกัด (มหาชน)' in txt:
            return 'PLT'
        if txt =='บริษัท พรูเด็นเชียล&#160;ประกันชีวิต (ประเทศไทย) จำกัด (มหาชน)' or 'Member' in txt:
            return 'PLT'
        else:
            return 'PLT' 
    except:
        return 'PLT'
print('JEDI') 
```

```python
df2['PAYER'].value_counts()
df2['PAYER']=df2['PAYER'].apply(payer)
df2['PAYER'].value_counts()
```

```python
def label(a):
    if a =='บริษัท พรูเด็นเชียล ประกันชีวิต (ประเทศไทย) จำกัด (มหาชน)':
        return 'PLT'
    else:
        a == 'PLT'
print('Jedi') 
```

```python
df2['PAYER'].value_counts()
df2['PAYER']=df2['PAYER'].apply(label) 
df2['PAYER'].value_counts()
```
